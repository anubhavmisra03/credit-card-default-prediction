```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
file_link = 'https://github.com/ybifoundation/Dataset/raw/main/Credit%20Default.csv'

df = pd.read_csv(file_link)

print("Dataset preview:")
print(df.head())
```

```python
print("\nShape of data:", df.shape)

print("\nColumn names:", df.columns)

print("\nSummary:")
print(df.describe())
```

```python
# New column create
df['Income_Loan_Diff'] = df['Income'] - df['Loan']

print("\nNew feature added!")
```

```python
plt.figure()

plt.plot(df['Income'][:50], label="Income")
plt.plot(df['Loan'][:50], label="Loan")

plt.legend()
plt.title("Income vs Loan Trend (First 50 Records)")
plt.show()
```

```python
target = df['Default']

features = df.drop(['Default'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    features, target, test_size=0.3, random_state=10
)
```

```python
from sklearn.tree import DecisionTreeClassifier

dt_model = DecisionTreeClassifier(max_depth=5)
```

```python
dt_model.fit(X_train, y_train)
```

```python
predictions = dt_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, classification_report

acc = accuracy_score(y_test, predictions)

print("\nAccuracy:", acc)

print("\nReport:\n")
print(classification_report(y_test, predictions))
```
